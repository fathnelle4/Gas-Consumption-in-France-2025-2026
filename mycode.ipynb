```python
# import 
import os
import requests
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, to_date, year
```

```python
spark = SparkSession.builder \
    .appName("GasPricesAnalysis") \
    .getOrCreate()

spark
```

```python


BASE_URL = "https://raw.githubusercontent.com/rvm-courses/GasPrices/main/"
FILES = [
    "Prix_2022_part1.csv",
    "Prix_2022_part2.csv",
    "Prix_2023.csv",
    "Prix_2024.csv",
    "Stations_2024.csv",
    "Services_2024.csv"
]

data_dir = "../data/raw"
os.makedirs(data_dir, exist_ok=True)

for file in FILES:
    url = BASE_URL + file
    path = os.path.join(data_dir, file)
    r = requests.get(url)
    with open(path, "wb") as f:
        f.write(r.content)

print("Téléchargement terminé ✅")
```
